```python
import numpy as np
from sklearn.linear_model import Perceptron
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report

X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([0, 1, 1, 0])

print("XOR Veri Kumesi:")
for xi, yi in zip(X, y):
    print(f"{xi} -> {yi}")
```

# Tek katmanlı perceptron

```python
perceptron = Perceptron(max_iter=1000, tol=1e-3, random_state=42)
perceptron.fit(X, y)

p_pred = perceptron.predict(X)
p_acc = accuracy_score(y, p_pred)

print("=== Tek Katmanli Perceptron ===")
print("Gercek Etiketler   :", y)
print("Tahmin Edilenler   :", p_pred)
print("Dogruluk (Accuracy):", p_acc)
print("\nSiniflandirma Raporu:\n")
print(classification_report(y, p_pred, zero_division=0))
```

# Çok katmanlı sinir ağı

```python
mlp = MLPClassifier(
    hidden_layer_sizes=(4,),
    activation='tanh',
    solver='lbfgs',
    max_iter=1000,
    random_state=42
)

mlp.fit(X, y)

m_pred = mlp.predict(X)
m_acc = accuracy_score(y, m_pred)

print("=== Cok Katmanli Sinir Agi (MLP) ===")
print("Gercek Etiketler   :", y)
print("Tahmin Edilenler   :", m_pred)
print("Dogruluk (Accuracy):", m_acc)
print("\nSiniflandirma Raporu:\n")
print(classification_report(y, m_pred, zero_division=0))
```
